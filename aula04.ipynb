## Conjuntos (set)

Um conjunto é uma coleção não ordenada de valores únicos.
1. Armazena apenas itens não duplicados (únicos).
2. Suporta operações matemáticas sobre conjuntos, como união e interseção.
3. Não é possível modificar os itens existentes, mas podemos adicionar novos elementos (conjuntos são mutáveis).
4. Suporta elementos de qualquer tipo, não ordenados e não indexados.
5. Pode conter apenas objetos imutáveis, como strings, ints, floats e tuplas de objetos também imutáveis.
6. São escritos com chaves {}.

```python
# Lista: Crie uma estrutura de repetição para fazer a leitura de 5 números inteiros e os armazene dentro de uma lista.
# Após a leitura, crie outra estrutura de repetição para somar todos os valores digitados

numeros = []
for i in range(5):
  n =int(input("digite um numero: "))
  numeros.append(n)

  soma = 0
  for numero in numeros:
    soma = soma + numero
  print(f" a soma dos numeros é :", soma )
```

```python
# 2. Dicionário: Crie um dicionário para armazenar o nome e a nota de 3 alunos, fazendo a leitura dos valores por meio de uma estrutura de repetição.
# Depois, crie uma nova estrutura de repetição para somar todas as notas e retornar a média

aluno = {}
for i in range(3):
  nome = input("digite o nome do aluno: ")
  nota = float(input("digite a nota do aluno: "))
  aluno[nome] = nota

soma = 0
for nota in aluno.values():
  soma += nota

media = soma/ len(aluno)


print ("media das notas : " , media )
```

```python
# 3. Matriz: Dada a matriz abaixo, construa uma estrutura de repetição para percorrer e somar todos os elementos da matriz

# matriz = np.array([[3, 4, 1],
#                    [3, 1, 5]])
```

```python
import numpy as np

matriz = np.array([[3,4,1],
                    [3,1,5]])

soma = 0

for linha in matriz :
  for elemento in linha :
    soma = soma + elemento

print("soma dos elementos",soma)
```
