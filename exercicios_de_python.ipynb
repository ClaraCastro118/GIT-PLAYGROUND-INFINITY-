```python
# 1.Análise de Strings: Peça ao usuário uma frase e informe: quantidade total de caracteres, quantidade de vogais, quantidade de consoantes e quantidade de espaços.

frase = input("Digite uma frase: ")
vogais = 0
consoantes = 0
espacos = 0
total_caracteres = len(frase)

for letra in frase.lower():
    if letra in "aeiou":
        vogais += 1
    elif letra == " ":
        espacos += 1
    elif letra.isalpha():
        consoantes += 1



```

```python
print("total de caracteres: ", total_caracteres)
print("quantidade de vogais: ", vogais)
print("quantidade de consoantes: " , consoantes)
print("quantidade de espaços: " , espacos)
```

```python
# 2.Contador Inteligente: Peça dois números (inicial e final). O programa deve imprimir todos os números entre eles. Se o primeiro número for maior, faça contagem decrescente; se for menor, faça contagem crescente.


num_inicial = int(input("Digite o número inicial: "))
num_final = int(input("Digite o número final: "))

if num_inicial < num_final:
    for num in range(num_inicial, num_final +1):
      print(num)
elif num_inicial > num_final:
    for num in range(num_inicial, num_final -1, -1):
      print(num)
else:
  print("Os dois números são iguais" , num_inicial)



```

```python
# 3.Validador de Senha: Peça uma senha ao usuário. A senha será válida apenas se possuir no mínimo 8 caracteres, pelo menos 1 número e pelo menos 1 letra.

senha = input("Digite uma senha: ")
tem_numero = False
tem_letra = False

for cacteres in senha:
  if cacteres.isnumeric():
    tem_numero = True
  elif cacteres.isalpha():
    tem_letra = True


if len(senha) >= 8 and tem_numero and tem_letra:
  print("Senha válida")
else:
  print("Senha inválida")


```

```python
#  4.Soma de Números Pares: Peça ao usuário 10 números, armazene em uma lista e mostre a soma apenas dos números pares e quantos pares foram digitados.


numeros = []

for i in range(10):
  num = int(input("Digite um número: "))
  numeros.append(num)

soma_pares = 0
quantidade_pares = 0

for numero in numeros:
  if numero % 2 == 0:
    soma_pares += numero
    quantidade_pares += 1

print("A soma dos números pares é: ", soma_pares)
print("A quantidade de números pares é: ", quantidade_pares)



# não consegui colocar bonitinho numero 1 numero 2 ...
# quando usei o f("") quebrou meu codigo



```

```python
# 7.Análise de Lista: Peça 10 números e mostre o maior número, o menor número, a média dos valores e quantos números são maiores que a média.


numeros = []


for i in range(10):
    num = float(input("Digite um número: "))
    numeros.append(num)

maior = max(numeros)
menor = min(numeros)
media = sum(numeros) / len(numeros)

contador = 0


for numero in numeros:
  if numero > media:
    contador += 1


print("maior numero é: ", maior)
print("menor numero é: ", menor)
print("a média dos valores é: ", media)
print("quantidade de numeros maiores que a media é :" , contador)

```

```python
# 8.Matriz Inteligente: Crie uma matriz 3x3 preenchida pelo usuário e mostre a soma de todos os valores, o maior número da matriz e a soma da diagonal principal.

matriz = []

soma_total = 0
maior = None
soma_diagonal = 0

for i in range(3):
  linha = []
  for j in range(3):
    num = int(input(f"Digite o valor para a posição [{i}][{j}]: "))
    linha.append(num)

    soma_total += num

    if maior is None or num > maior:
      maior = num

    if i == j:
      soma_diagonal += num

    matriz.append(linha)

print("Matriz:")
for linha in matriz:
  print(linha)



print("soma total: ", soma_total)
print("maior numero: ", maior)
print("soma da diagonal principal: ", soma_diagonal)



```

```python
#9. Função de Número Primo: Crie uma função chamada eh_primo(numero) que verifique se um número é primo. Depois peça um número ao usuário e utilize a função.


def eh_primo(numero):
  if numero <= 1:
     return False

  for i in range(2, numero):
    if numero % i == 0:
      return False
  return True

numero = int(input("Digite um número: "))

if eh_primo(numero):
  print(f"{numero} é primo")
else:
    print(f"{numero} não é primo")




```

```python
# 10. Sistema de Notas: Peça o nome de 5 alunos e duas notas para cada um. Calcule a média e mostre a situação (Aprovado, Recuperação ou Reprovado).

#5 alunos
# notas
# media
# aprovado reprovado recuperação

for i in range(5):
  nome = input("Digite o nome do aluno: ")

  nota01 = float(input("Digite a primeira nota: "))
  nota02 = float(input("Digite a segunda nota: "))

  media = (nota01 + nota02) / 2

  if media >= 7:
    situacao = "Aprovado"
  elif media >= 5:
    situacao = "Recuperação"
  else:
    situacao = "Reprovado"

  print(f"Aluno: {nome}")
  print(f"Média: {media}")
  print(f"Situação: {situacao}")
  print("----------------------")

```

```python
# 11.Jogo de Adivinhação: Gere um número aleatório entre 1 e 100. O usuário deve tentar adivinhar. O programa informa se o palpite é muito alto, muito baixo ou correto, e ao final mostra o número de tentativas.

#nao entendi ?

#teria que ter um numero de 1 a 100 de forma aleatoria ? mas como ? random

import random

numero_secreto = random.randint(1, 100)
tentativas = 0

while True:
  palpite = int(input("Tente adivinhar o número secreto (entre 1 e 100:"))
  tentativas += 1

  if palpite < numero_secreto:
    print("Tente um número maior")
  elif palpite > numero_secreto:
    print("Tente um número menor")
  else:
    print ("Parabens! Voce acertou!")
    print(f"Numero de tentativas: {tentativas}")
    break



```

```python
#  12. Analisador de Palavras: Peça uma frase e mostre a palavra mais longa, a palavra mais curta, a quantidade de palavras e quantas começam com vogal.

frase = input("Digite uma frase: ")

#como divido as palavras na frase ? slpit

palavras = frase.split()
quantidade = len(palavras)

maior = palavras[0]
menor = palavras[0]

vogais = 0


for palavra in palavras:
  if len(palavra) > len(maior):
    maior = palavra

  if len(palavra) < len(menor):
    menor = palavra

  if palavra[0].lower() in "aeiou":
    vogais += 1

  print("quantidade de palavras: ", quantidade)
  print("maior palavra: ", maior)
  print("menor palavra: ", menor)
  print("quantidade de palavras que começam com vogal: ", vogais)



```
